```python
import pandas as pd
```

```python
def concrete_stress_comp(fck:int):
    """
    Returns the compressive stress in concrete
    """
    cbc_df=pd.read_csv("Concrete stress in compression.csv")
    cbc_df=cbc_df[:10]
    cbc_df=cbc_df.set_index("fck")
    cbc=cbc_df.loc[fck,'cbc']
    return cbc
```

```python
concrete_stress_comp(30)
```
